* **세부 요구사항**
    - 데이터셋을 전처리 합니다.

#### 1) 모델이 요하는 구조의 폴더 생성

```python
import glob
import random
import shutil
import numpy as np

from keras.utils import load_img,img_to_array
from keras.utils import image_dataset_from_directory
```

```python
## image_dataset_from_directory 사용하기 위하여 폴더 생성

# 경로
tr_data = '/content/train'
te_data = '/content/test'
val_data = '/content/val'
# 폴더가 없으면 생성
if not os.path.exists(tr_data):
    os.makedirs(tr_data)

if not os.path.exists(te_data):
    os.makedirs(te_data)

if not os.path.exists(val_data):
    os.makedirs(val_data)

# 생성 확인
print(os.path.exists(tr_data))
print(os.path.exists(te_data))
print(os.path.exists(val_data))
```

#### 2) 각 폴더에 이미지 데이터 옮기기

```python
# Keras image_dataset_from_directory 하위 폴더 생성

# 하위 폴더 생성
class_names = ['my','other']

for cn in class_names:
    temp = os.path.join(tr_data,cn)

    if not os.path.exists(temp):
        os.makedirs(temp)

    # 폴더 생성 확인
    print(os.path.exists(temp))

for cn in class_names:
    temp = os.path.join(te_data,cn)

    if not os.path.exists(temp):
        os.makedirs(temp)

    # 폴더 생성 확인
    print(os.path.exists(temp))

for cn in class_names:
    temp = os.path.join(val_data,cn)

    if not os.path.exists(temp):
        os.makedirs(temp)

    print(os.path.exists(temp))
```

```python
# 내 얼굴 데이터가 있는 폴더 경로의 파일 전체를 정렬하여 리스트화

img_list_my = sorted(glob.glob('/content/my_face/*'))

# 다른 사람 얼굴 데이터가 있는 폴더 경로의 파일 전체를 정렬하여 리스트화
img_list_other = sorted(glob.glob('/content/other_face/*'))

# 이미지 갯수 확인
len(img_list_my),len(img_list_other)
```

```python
# 얼굴 데이터를 training set, test set으로 분할하기 위한 작업
# 재현성을 위한 난수 고정
random.seed(2024)
random.shuffle(img_list_my)
random.shuffle(img_list_other)

img_list_my[:5],img_list_other[:5]
```

```python
## Test set의 비율 설정
test_size = 0.2

## Validation set의 비율 설정
val_size = 0.2

## 각 리스트를 Training, Validation, Test set으로 분할
for i_l, class_name in zip([img_list_my, img_list_other], class_names):
    # 리스트 길이 제한 설정 (예: 11000개까지만 사용)
    list_len = 11000

    # 데이터 분할을 위한 인덱스 설정
    split_test_idx = int(list_len * (1 - test_size))
    split_val_idx = int(split_test_idx * (1 - val_size))

    # Training, Validation, Test set 분할
    list_tr = i_l[:split_val_idx]
    list_val = i_l[split_val_idx:split_test_idx]
    list_te = i_l[split_test_idx:list_len]

    # 파일 복사 함수 정의
    def copy_files(file_list, dst_folder, class_name):
        for file_path in file_list:
            f_name = file_path.split('/')[-1]
            shutil.copy(src=file_path, dst=os.path.join(dst_folder, class_name, f_name))
            print(f'파일 이동 완료: {f_name} → {dst_folder}/{class_name}/')

    # 각 분할 데이터에 대해 파일 복사
    copy_files(list_tr, tr_data, class_name)
    copy_files(list_val, val_data, class_name)
    copy_files(list_te, te_data, class_name)
```

```python
# 결과 확인
print(f"Training set의 'my' 이미지 수: {len(os.listdir(os.path.join(tr_data, 'my')))}")
print(f"Validation set의 'my' 이미지 수: {len(os.listdir(os.path.join(val_data, 'my')))}")
print(f"Test set의 'my' 이미지 수: {len(os.listdir(os.path.join(te_data, 'my')))}")

print(f"Training set의 'other' 이미지 수: {len(os.listdir(os.path.join(tr_data, 'other')))}")
print(f"Validation set의 'other' 이미지 수: {len(os.listdir(os.path.join(val_data, 'other')))}")
print(f"Test set의 'other' 이미지 수: {len(os.listdir(os.path.join(te_data, 'other')))}")
```

#### (1) UltraLytics YOLO-cls 모델 선택

```python
from ultralytics import YOLO
```

#### (2) UltraLytics YOLO-cls 모델 학습

* **세부 요구사항**
    - 선택한 UltraLytics YOLO-cls 모델로 학습을 진행합니다.
    - [UltraLytics YOLO 학습 명령어 링크](https://docs.ultralytics.com/modes/train/#train-settings)

```python
# YOLO-cls X 모델 로드
model = YOLO('yolov8n-cls.pt')  # YOLO-cls X 모델 사전학습 가중치
model.model.nc = 2  # 클래스 수를 2로 설정 ('my'와 'other' 클래스)
```

```python
from torchvision import datasets

# 데이터셋 확인
train_dataset = datasets.ImageFolder('/content/train')
print("클래스:", train_dataset.classes)
print("클래스 레이블 매핑:", train_dataset.class_to_idx)
print("데이터셋 내 이미지 수:", len(train_dataset))
```

```python
import torch
print("CUDA Available:", torch.cuda.is_available())
print("CUDA Device Count:", torch.cuda.device_count())
print("Current CUDA Device:", torch.cuda.current_device())
```

```python
# CUDA 런타임 오류 디버깅 활성화
os.environ['CUDA_LAUNCH_BLOCKING'] = '1'
model.train(data='/content/',
            lr0=0.01,               # 학습률 0.01
            dfl=0.5,                # 위치정도에 대해 민감하게 함
            epochs=30,
            imgsz=224,
            weight_decay=0.0005,    # 규제 추가
            momentum = 0.9,         # 모멘텀 추가
            lrf =0.02,              # 학습이 진행됨에 따라 학습률을 서서히 낮추는 함수
            warmup_epochs=12,       # 초기 epochs를 점차 증가시키는 함수
            batch=8,device='cuda:0')
```

#### (3) UltraLytics YOLO-cls 추론

* **세부 요구사항**
    - 학습이 완료되면 추론을 진행합니다.
    - [UltraLytics YOLO 추론 명령어 링크](https://docs.ultralytics.com/modes/predict/)

```python
# 통합 폴더 생성
merged_dir = "/content/merged_test"
os.makedirs(merged_dir, exist_ok=True)

# te_data 하위 폴더(my와 other)에서 이미지 파일을 merged_dir로 복사
for folder in ["/content/test/my", "/content/test/other"]:
    for file_name in os.listdir(folder):
        src_path = os.path.join(folder, file_name)
        dst_path = os.path.join(merged_dir, file_name)
        if os.path.isfile(src_path):  # 파일만 복사
            shutil.copy(src_path, dst_path)
```

```python
results = model.predict(source=merged_dir)
print(results)
```
